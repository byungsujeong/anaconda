```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
y = 200+np.random.randn(100)
```

```python
x=[x for x in range(len(y))]
```

```python
plt.plot(x,y,'-')
plt.fill_between(x,y,195,where=(y>195),facecolor='g', alpha=0.6)
plt.title('graph')
plt.show()
```
